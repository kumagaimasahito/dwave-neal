```python
import neal
import numpy as np
from utils import load_clustering_ising
import matplotlib.pyplot as plt
```

## Easy problem

```python
sampler = neal.SimulatedAnnealingSampler()
h = {'a': 0.0, 'b': 0.0, 'c': 0.0}
J = {('a', 'b'): 1.0, ('b', 'c'): 1.0, ('a', 'c'): 1.0}
sampleset = sampler.sample_ising(h, J, num_reads=1)
```

```python
sampleset.best_solution
```

## Clustering

```python
n_clusters = 3
n_points = 8
```

```python
lci = load_clustering_ising(n_clusters, n_points, random_state=0)
Jm, hm = lci.get()
```

```python
sampler = neal.SimulatedAnnealingSampler()
sampleset = sampler.sample_ising(hm, Jm, num_reads=1)
```

```python
sampleset.best_solution
```

```python
sampleset.first.energy
```

```python
plt.plot(sampleset.trajectory["energies"])
plt.xlabel("Time steps")
plt.ylabel("Energy")
plt.show()
```

```python
# 
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6.4, 4.8))

# 
ax1.plot(sampleset.trajectory["states"].T[0])
ax1.set_ylabel("Spin #1")

ax2.plot(sampleset.trajectory["states"].T[1])
ax2.set_ylabel("Spin #2")

ax3.plot(sampleset.trajectory["states"].T[2])
ax3.set_ylabel("Spin #3")
ax3.set_xlabel("Time steps")

# 
plt.tight_layout()

# 
plt.show()
```

```python
# 
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6.4, 4.8))

# 
ax1.plot(sampleset.trajectory["p_flip"].T[0])
ax1.set_ylim(-0.1, 1)
ax1.set_ylabel("$p_{flip}$")

ax2.plot(sampleset.trajectory["p_flip"].T[1])
ax2.set_ylim(-0.1, 1)
ax2.set_ylabel("$p_{flip}$")

ax3.plot(sampleset.trajectory["p_flip"].T[2])
ax3.set_ylim(-0.1, 1)
ax3.set_ylabel("$p_{flip}$")
ax3.set_xlabel("Time steps")

# 
plt.tight_layout()

# 
plt.show()
```
